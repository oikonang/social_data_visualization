*Q1. Create your own version of Figure 3-1, but change the values, axes-labels, and the plot title. Just add the plot to the bottom of the notebook you've already been working for Part 3 of today's lecture.*

```python
from matplotlib import pyplot as plt
import matplotlib

# Define the figure object and add a subplot to it
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)

# Define x and y axis
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016]
knowledge_points = [10, 23.4, 25, 39.5, 40, 45, 57]

# Fit the axis to the plot
ax.plot(years, knowledge_points, color='black', marker='+', linestyle='solid')

# Fix the xaxis bug of not presenting the whole numbers
x_formatter = matplotlib.ticker.ScalarFormatter(useOffset=False)
ax.xaxis.set_major_formatter(x_formatter)

ax.set_title("Knowledge over time")
ax.set_ylabel("Knowledge Points")
ax.set_xlim(2010,2016)

plt.show()
```

*Q2. Create your own version of Figure 3-2, but change the values, axes-labels, and the plot title. Just add the plot to the bottom of the notebook you've already been working for Part 3 of today's lecture.*

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

# Assign x and y axis
courses = ["Machine Learning", "Social Data", "Cognitive Science", "Big Data", "UX Engineering"]
score = [2, 7, 0, 12, 10]

# Give indeces to courses
x_pos = np.arange(len(courses))

# Assign the size of the plot before declaring the plot
plt.figure(figsize= (10,3))

# Give the x and y values to the barplot
plt.bar(x_pos, score, align='center', alpha=0.5)

# Match the indeces with the correct courses and assign them to the x-axis
plt.xticks(x_pos, courses)

# Assign all possible scores that a student can achieve
plt.yticks([-2, 0, 2, 4, 7, 10, 12])

# Give labels and titles
plt.ylabel("Grades")
plt.title("My Personal Score System")

# Show the barplot
plt.show()
```

*Q4. Create your own version of Figure 3-9 and add to the notebook. Explain in your own words why 3-9 is a better representation of the data than 3-8.*

```python
# Assign x and y axis
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

# Assign the size of the plot before declaring the plot
plt.figure(figsize= (6,4))

# Give the x and y values to the scatterplot
plt.scatter(test_1_grades, test_2_grades)

# Assign number of mentions to the y-axis and x-axis accordingly
plt.yticks(range(60,110,10))
plt.xticks(range(70,110,10))

# Assign equal measure between the two axis 
plt.axis("equal")

# Add titles and labels
plt.title("Axes Are Comparable after the 'equal' axis assignment")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")
plt.show()
```
